```python
# ==============================
# Multiple Linear Regression - Home Price
# ==============================

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from google.colab import files

# Upload CSV file manually
uploaded = files.upload()

# Load dataset
df = pd.read_csv(list(uploaded.keys())[0])

print("Dataset Preview:")
display(df)

# ==============================
# Data Preprocessing
# ==============================

# Fill missing bedrooms with median
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())

print("\nAfter Filling Missing Values:")
display(df)

# ==============================
# Model Training
# ==============================

X = df.drop('price', axis='columns')
y = df['price']

model = LinearRegression()
model.fit(X, y)

print("\nModel Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)

# ==============================
# Prediction
# ==============================

# Example: 3000 sqft, 3 bedrooms, 40 years old
prediction = model.predict([[3000, 3, 40]])

print("\nPredicted Price for (3000 sqft, 3 bedrooms, 40 yrs):")
print(prediction[0])
```

```python
# ==============================
# Simple Linear Regression - Area vs Price
# ==============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from google.colab import files

# Upload CSV file
uploaded = files.upload()

# Load dataset
df = pd.read_csv(list(uploaded.keys())[0])

print("Dataset Preview:")
display(df)

# ==============================
# Visualization
# ==============================

plt.figure()
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df['area'], df['price'], marker='+')
plt.title("Area vs Price")
plt.show()

# ==============================
# Model Training
# ==============================

X = df[['area']]
y = df['price']

model = LinearRegression()
model.fit(X, y)

print("\nModel Coefficient (m):", model.coef_[0])
print("Model Intercept (b):", model.intercept_)

# ==============================
# Predictions
# ==============================

# Predict for 3300 sq ft
pred_3300 = model.predict([[3300]])
print("\nPredicted price for 3300 sq ft:", pred_3300[0])

# Predict for 5000 sq ft
pred_5000 = model.predict([[5000]])
print("Predicted price for 5000 sq ft:", pred_5000[0])
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from google.colab import files

# Upload file
uploaded = files.upload()
df = pd.read_csv(list(uploaded.keys())[0])

print("Original Data:")
display(df)

# -------------------------
# Convert word numbers to digits (FULL dictionary)
# -------------------------

word_to_num = {
    'zero':0, 'one':1, 'two':2, 'three':3, 'four':4,
    'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':9,
    'ten':10, 'eleven':11, 'twelve':12, 'thirteen':13,
    'fourteen':14, 'fifteen':15
}

df['experience'] = df['experience'].replace(word_to_num)

# Convert to numeric safely
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')

# -------------------------
# Handle Missing Values
# -------------------------

df['experience'] = df['experience'].fillna(df['experience'].median())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median())
df['interview_score(out of 10)'] = df['interview_score(out of 10)'].fillna(df['interview_score(out of 10)'].median())

print("After Cleaning:")
display(df)

# -------------------------
# Model Training
# -------------------------

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

model = LinearRegression()
model.fit(X, y)

# -------------------------
# Predictions
# -------------------------

print("2 yr exp, 9 test, 6 interview:", model.predict([[2,9,6]])[0])
print("12 yr exp, 10 test, 10 interview:", model.predict([[12,10,10]])[0])
```

```python
# ==============================
# 1000 Companies Profit Prediction
# ==============================

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from google.colab import files

# Upload 1000_companies.csv
uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])

print("Dataset Preview:")
display(df)

# ------------------------------
# Encoding Categorical Variable (State)
# ------------------------------

df = pd.get_dummies(df, drop_first=True)

# Define features and target
X = df.drop('Profit', axis=1)
y = df['Profit']

# Build model
model = LinearRegression()
model.fit(X, y)

# ------------------------------
# Prediction
# ------------------------------

# Create input in same order as training columns
input_data = pd.DataFrame(
    [[91694.48, 515841.3, 11931.24, 1, 0]],  # Florida encoded
    columns=X.columns
)

pred_profit = model.predict(input_data)

print("Predicted Profit:", pred_profit[0])
```
